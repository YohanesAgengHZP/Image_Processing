```python
import skimage
from skimage import data
import matplotlib.pyplot as plt
```

```python
camera = data.camera()
```

```python
print("Tipe image - camera: ", type(camera))
```

```python
plt.imshow(camera, cmap='gray')
```

```python
plt.show()
```

```python
camera
```

```python
num_rows, num_cols = camera.shape
print("Jumlah row:", num_rows)
print("Jumlah kolom:", num_cols)
```

```python
# Copy gambar
camera_bright = camera.copy()
# Nilai brightnes
b = 50
for i in range(num_rows):
    for j in range(num_cols):
        temp = camera[i][j] + b
        if (temp > 255):
            camera_bright[i][j] = 255
        elif(temp < 0):
            camera_bright[i][j] = 0
        else:
            camera_bright[i][j] = temp
```

```python
camera_bright
```

```python
f, axarr = plt.subplots(1,2, figsize=(12,12))
axarr[0].imshow(camera, cmap='gray')
axarr[1].imshow(camera_bright, cmap='gray')
```
