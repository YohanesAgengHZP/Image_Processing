```python
import skimage
from skimage import data
import matplotlib.pyplot as plt
from PIL import Image
```

```python
camera = data.camera()
```

```python
print("Tipe image - camera: ", type(camera))
```

```python
plt.imshow(camera, cmap='gray')
```

```python
camera
```

```python
num_rows, num_cols = camera.shape
print("Jumlah row:", num_rows)
print("Jumlah kolom:", num_cols)
```

```python
# Copy gambar
camera_biner = camera.copy()
# Iterasi tiap elemen, ganti nilai menjadi negatif nya
for i in range(num_rows):
    for j in range(num_cols):
        # threshold = 128
        if(camera[i][j] < 100):
            camera_biner[i][j] = 0
        else:
            camera_biner[i][j] = 1
```

```python
camera_biner
```

```python
f, axarr = plt.subplots(1,2, figsize=(12,12))
axarr[0].imshow(camera, cmap='gray')
axarr[1].imshow(camera_biner, cmap='gray')
```
