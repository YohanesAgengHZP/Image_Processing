```python
import skimage
from skimage import data
import matplotlib.pyplot as plt
```

```python
coins = data.coins()
```

```python
print("Tipe image - coins: ", type(coins))
```

```python
plt.imshow(coins, cmap='gray');
```

```python
plt.show()
```

```python
coins
```

```python
num_rows, num_cols = coins.shape
print("Jumlah row:", num_rows)
print("Jumlah kolom:", num_cols)
```

```python
coins_negatif = coins.copy()
# Iterasi tiap elemen, ganti nilai menjadi negatif nya
for i in range(num_rows):
    for j in range(num_cols):
        coins_negatif[i][j] = 255 - coins[i][j]
```

```python
coins_negatif
```

```python
plt.imshow(coins_negatif, cmap='gray');
```

```python
f, axarr = plt.subplots(1,2, figsize=(15,15))
axarr[0].imshow(coins, cmap='gray')
axarr[1].imshow(coins_negatif, cmap='gray')
```
