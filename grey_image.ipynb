```python
# Pakai library skimage
import skimage
# Ambil data sample
from skimage import data
# Pakai library plotting gambar
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Ambil gambar dari sample data di paket SKImage
astro = data.astronaut()
```

```python
print("Tipe image - astro: ", type(astro))
```

```python
# Plot gambar asli
plt.imshow(astro, cmap='gray')
```

```python
plt.show()
```

```python
# Lihat data array gambar (Numpy Array)
astro
```

```python
num_rows, num_cols, num_rgb = astro.shape
print('Jumlah rows: ', num_rows)
print('Jumlah cols: ', num_cols)
print('Jumlah elemen/pixel: ', num_rgb)
```

```python
astro_gray = np.zeros((num_rows,num_cols))
type(astro_gray)
```

```python
for i in range(num_rows):
    for j in range(num_cols):
        temp = 0.299*astro[i][j][0] + 0.587*astro[i][j][1] + 0.144*astro[i][j][2]
        astro_gray[i][j] = temp
```

```python
astro_gray.astype(np.uint8)
```

```python
f, axarr = plt.subplots(1,2, figsize=(12,12))
axarr[0].imshow(astro, cmap='gray')
axarr[1].imshow(astro_gray.astype(np.uint8), cmap='gray')
```
