```python
import skimage
from skimage import data
import matplotlib.pyplot as plt
import numpy as np
horse = data.horse()
```

```python
type(horse)
```

```python
plt.imshow(horse, cmap='gray')
```

```python
plt.show()
```

```python
num_rows, num_cols = horse.shape
horse_negatif = np.ones([num_rows,num_cols], dtype=bool)
horse_negatif
```

```python
for i in range(num_rows):
    for j in range(num_cols):
        horse_negatif[i][j] = not horse[i][j]

horse_negatif
```

```python
f, axarr = plt.subplots(1,2, figsize=(20,20))
axarr[0].imshow(horse, cmap='gray')
axarr[1].imshow(horse_negatif, cmap='gray')
```
